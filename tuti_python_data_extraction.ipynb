```python
import matplotlib.pyplot as plt
```

```python
data_frame_01["Box Office"].plot(kind='line', title="Box Office over Budget")
plt.ylabel("Budget")
plt.xlabel("Box Office")
```

```python
films_per_year = data_frame_01.groupby("Year").size()
films_per_year = data_frame_01.groupby("Year")["Actor"].count()
```

```python
films_per_year.plot(kind='line', title="Number of Films Released Per Year", figsize=(10,5))
plt.xlabel("Year")
plt.ylabel("Number of Films")
plt.grid(True)
```

```python
data_frame_01.plot(kind='scatter', x='Budget', y='Box Office', title="Budget vs Box Office")
```

```python
data_frame_01["Actor"].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title="Actor Appearance Share")
plt.ylabel("")
```

```python
data_frame_01["Budget"].plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8, 8),
    title='Percentage of Budget Covered by Each Film',
    legend=False
)
plt.ylabel("") 
plt.tight_layout()
```
