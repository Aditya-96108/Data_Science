```python
class Player : 
    def __init__(self, name, team):
        self.name=name
        self.team=team
    def describe(self):
            return f"{self.name} plays for {self.team}"
```

```python
p1=Player("Rohit Sharma","Mumbai Indians")
print(p1.describe())
```
